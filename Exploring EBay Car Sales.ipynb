Regarding "price" and "odometer_km", in practice they are numerical data, but in this dataset they are treated as "string" data. Thats because it contains non-numerical characters, such as '$' and 'km'. So, we clear these variables and convert them to "float" type

```python
autos['price'] = autos['price'].str.replace('$','').str.replace(',','').astype(float)
```

```python
autos['odometer_km'] = autos['odometer_km'].str.replace('km','').str.replace(',','').astype(float)
```

This next step is focused on analyzing the mean price and mean mileage across the top 20 brands by sale.

First step involves selecting the index of those top 20 brands.

```python
indexes = autos['brand'].value_counts().head(20).index
```

Then, using dictionaries, we get the mean of each brand.

To do this, we iterate for those 20 brands, locate them in the dataset, calculate their respective means and assign each to their specific brand.

```python
mean_price = {}
mean_mileage ={}

for i in indexes:
    brand_p = autos.loc[autos['brand'] == i, 'price']
    mean_p = brand_p.mean()
    mean_price[i] = mean_p
    
    brand_m = autos.loc[autos['brand'] == i, 'odometer_km']
    mean_m = brand_m.mean()
    mean_mileage[i] = mean_m
    
```

Now we convert them into one DataFrame

```python
df_price = pd.Series(mean_price)
df_mileage = pd.Series(mean_mileage)
```

```python
df = pd.DataFrame(df_price, columns = ['price'])
```

```python
df['mileage'] = df_mileage
```

```python
df
```

From these results, we can see that some brands are still more valuable than others, even with higher mileage. 

BWM has a higher mean mileage, but higher prices than Volkswagen.

Also we observe that brands like Volvo, Smart and Fiat have a lower price/mileage ratio, indicating that their cars are not so valuable as their counterparts, or their value deprecates a lot with higher mileage.

The next step consists in dropping columns with the same data in almost all of the series.

To filter them, we select the main repeated value (to filter the outliers) and drop the columns. By doing this, we guarantee that all of our future analysis will consist of vehicles with the same information regarding "Seller" and "offer_type"

```python
autos['seller'] = autos['seller'] == 'private'
autos['offer_type'] = autos['offer_type'] == 'Angebot'
```

Now we drop the columns containing the same value accross all dataset

```python
autos.drop(columns = ['seller','offer_type','nr_of_pictures'], inplace = True)
```

Now we treat the 'ad_created' variable. Its notable that its a "datetime" type, but its "hour" part is all 00:00. So we will convert it to a string type consisting of yyyymmdd (year-month-date).

To do this, we remove the '-' and get only the first 8 character from the string.

```python
autos['ad_created'] = autos['ad_created'].str.replace('-','').str[:8]
```

Now, to finish the cleaning of the data, we will translate all the german words to english. We will use the method 'replace' from Pandas.

```python
autos['vehicle_type'].replace({'kleinwagen':'small_car','cabrio':'convertible','andere':'other'}, inplace = True)
```

```python
autos['gearbox'].replace({'mannuel':'manual','automatik':'automatic'}, inplace = True)
```

```python
autos['fuel_type'].replace({'benzin':'gasoline','andere':'other'}, inplace = True)
```

```python
autos['unrepaired_damage'].replace({'nein':'no','ja':'yes'}, inplace = True)
```

```python
autos['model'].replace({'andere':'other'},inplace = True)
```
